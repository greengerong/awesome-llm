# chromadb

```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection(name="my_collection")

collection.add(
    documents=["ai技术学习文档", "这是一篇关于牛肉介绍", "关于番茄炒蛋的做法"],
    metadatas=[{"source": "doc1"}, {"source": "doc2"}, {"source": "doc3"}],
    ids=["id1", "id2", "id3"]
)
```

```python

results = collection.query(
    query_texts=["请选择一款食物?"],
    n_results=2
)

print(results)
```
